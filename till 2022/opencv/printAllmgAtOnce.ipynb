```python
import cv2
from matplotlib import pyplot as plt
from numpy import *
img1=cv2.imread('lena.jpg',1)
img2=cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
img3=cv2.adaptiveThreshold(img2,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
img4=cv2.adaptiveThreshold(img2,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)
titles=['original','gray','mean','gussian']
image=[img1,img2,img3,img4]
for i in range(4):
    plt.subplot(2,2,i+1), plt.imshow(image[i],'gray')
    plt.title(titles[i])
plt.show()
```
