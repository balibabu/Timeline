```python
import cv2
from numpy import *
from matplotlib import pyplot as plt
img=cv2.imread('lena.jpg',1)
b,g,r=cv2.split(img)
plt.hist(b.ravel(),256,[0,256])
plt.hist(g.ravel(),256,[0,256])
plt.hist(r.ravel(),256,[0,256])
plt.show()
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
print(b.ravel())
```
