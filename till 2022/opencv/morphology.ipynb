```python
import cv2
from numpy import *
from matplotlib import pyplot as plt
img=cv2.imread('smarties.png',0)
_,mask=cv2.threshold(img,220,255,cv2.THRESH_BINARY_INV)
kernal=ones([3,3],uint8)
dilation=cv2.dilate(mask,kernal,iterations=2)
erosion=cv2.erode(mask,kernal,iterations=2)
opening=cv2.morphologyEx(mask,cv2.MORPH_OPEN,kernal)
closing=cv2.morphologyEx(mask,cv2.MORPH_CLOSE,kernal)
titles=['original','mask','dilation','erosion','opening','closing']
image=[img,mask,dilation,erosion,opening,closing]
for i in range(6):
    plt.subplot(2,3,i+1), plt.imshow(image[i],'gray')
    plt.title(titles[i])
plt.show()
cv2.imshow('mask',mask)
cv2.imshow('testing',closing)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
