```python
class Obj:
    def __init__(self, v=None):
        self.set_value(v)
    def set_value(self, v):
        self._v = v
        self._content = v
    @property
    def content(self):
        return self._content
    def _compare_repeat(self, a, b, count = 0):
        if a.v == b.v :
            return makeNew(a,b,a)
        elif a.v == "R" and b.v == "P":
            return makeNew(a,b,b)
        elif a.v == "R" and b.v == "S":
            return makeNew(a,b,a)
        elif a.v == "P" and b.v == "S":
            return makeNew(a,b,b)
        else:
            if count > 2:
                raise IOError("Err")
            return self._compare_repeat(b, a, count + 1)
    def set_objects(self, objs: list):
        self._objs = objs
        self._v = self._evalualt(objs).v
        self._content = "".join([r.content for r in self._objs])
    def _evalualt(self, objs):
        if len(objs) == 2:
            return self._compare_repeat(objs[0], objs[1])
        return self._compare_repeat(self._evalualt(objs[: int(len(objs)/ 2)]), self._evalualt(objs[int(len(objs)/ 2): ]))
    def set_string(self, string):
        objs = []
        for i in string:
            objs.append(Obj(i))
        self.set_objects(objs)
    @property
    def v(self):
        if self._v:
            return self._v
        raise IOError("v not calculated")
    @property
    def vcontent(self):
        return "".join([r.v for r in self._objs])
```

```python
from ListDB import ListDB
def makeNew(a, b, r):
    obj = Obj()
    obj._content = a.content + b.content
    obj._v = r.v
    return obj
```

```python
def makeObj(letter, nr):
    objs = []
    for i in range(nr):
        objs.append(Obj(letter))
    return objs
def makeObjs(inp):
    aas = inp.split()
    aam = {}
    for a in aas:
        aam[a[-1]] = int(a[:-1])
    res = []
    for k, v in aam.items():
        res += makeObj(k, v)
    return res
def mappv(objs):
    count = {"R":[], "P": [], "S":[]}
    for o in objs:
        count[o.v].append(o)
    return count
def get_arrangement(pattern, values):
    res = []
    for p in pattern:
        res.append(values[p].pop())
    return res
def getNextLevel(objs):
    values = mappv(objs)
    r, p, s = [len(x) for x in  values.values()]
    res = ""
    if p >= r:
        res += "RP"* r
        res += "P" * (p-r)
        res += "S" * s
    else:
        res += "RP"* p
        rl = r-p
        res = "R" * rl + res + "S"* s
    tot = get_arrangement(res, values)
    lvl = ListDB.reshape(tot, (len(tot)/2, 2))
    combins = []
    for el in lvl:
        ob = Obj()
        ob.set_objects(el)
        combins.append(ob)
    return combins
```

```python

# calling and example

res = makeObjs("11R 3P 2S") # round 0
ob = Obj()
ob.set_objects(res)
print("round 0")
print(ob.v, ob.content) # ob.v :- who wins with current settings, ob.content :- current setting
print()

res = getNextLevel(res) # round 1 // set in such a way that more R's are removed
ob = Obj()
ob.set_objects(res)
print("round 1")
print(ob.v, ob.content) # next round winner and settings
print()

res = getNextLevel(res) # go to the next round
ob = Obj()
ob.set_objects(res)
print("round 2")
print(ob.v, ob.content) # next round winner and settings
print()
```
