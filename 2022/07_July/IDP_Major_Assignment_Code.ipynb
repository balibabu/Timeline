# Import Modules

```python
# import required headers

import pandas as pd
import numpy as np
from math import ceil
import matplotlib.pyplot as plt
import seaborn as sb
```

# 1. Reading the CSV File

```python
# read the CSV file
df = pd.read_csv('winequality-white.csv', sep = ';')
df.head()
```

```python
df.info()
```

# 2. Group Attributes into buckets and plot histogram

```python
# group by attribute and plot histogram 

df.hist(bins=30,figsize=(10,10))
plt.show()
```

# 3. Mean, Median, Mode and Variance

```python
# find mean
print("Mean")
df.mean()
```

```python
# find median 
print("Median")
df.median()
```

```python
# find mode
print("Mode")
df.mode()
```

```python
# variance
print("Variance")
df.var()
```

# 4. Covariance

```python
# covariance
df.cov()
```

# 5. Corelation Matrix

```python
# corelation matrix
df_corr = df.corr()
df_corr
```

# 6. Corelation of attributes ( similarity and disimilarity)

```python
# remove data redundancy
mask = np.triu(np.ones_like(df_corr, dtype=np.bool))

# corelation heatmap to find similarity and disimarity between paired attributes
plt.figure(figsize=[18,7])
sb.heatmap(df.corr(),mask=mask, annot=True)
plt.show()
```

# 6a. Most Corelated Attributes

```python
# refer heat map above

# sharing most similarity
print("DENSITY & RESIDUAL SUGAR share most similarity i.e POSITIVE CORELATION")

# sharing most disimilarity
print("ALCOHOL & DENSITY share most dissimilarity i.e NEGATIVE CORELATION")

```

# 6b. Attribute corelated with Output labels

```python
attr = []

# find output label ( sharing multi corelation)
for i in range(len(df_corr.keys())):
  for j in range(len(df_corr)):
    if abs(df_corr.iloc[i,j]) > 0.7:
      attr = df_corr.columns[i]

print("OUTPUT LABEL : ", attr)

# after referring to heatmap
print("MAXIMUM SIMILARITY (positive corealtion) with OUTPUT LABEL  i.e ", attr, " : ", "ALCOHOL")
print("MAXIMUM DISIMILARITY (negative corealtion) with OUTPUT LABEL i.e ", attr, " : ", "DENSITY")
```
